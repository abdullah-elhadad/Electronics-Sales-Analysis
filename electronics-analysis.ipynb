```python
top_products_sales = df.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
ax = sns.barplot(x=top_products_sales.index, y=top_products_sales.values, palette='coolwarm')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Total Sales')
plt.title('Top 10 Products by Sales', fontsize=16, color='darkblue')

ax.bar_label(ax.containers[0], fmt='%.0f', padding=5)

plt.show()
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
basket = df.groupby('Order ID')['Product'].apply(list)
basket = basket[basket.map(len) > 1] 

te = TransactionEncoder()
te_ary = te.fit(basket).transform(basket)
basket_df = pd.DataFrame(te_ary, columns=te.columns_)


frequent_itemsets = apriori(basket_df, min_support=0.01, use_colnames=True) 
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)


rules.sort_values('lift', ascending=False).head(5)
```

```python
important_cols = ['antecedents', 'consequents', 'support', 'confidence', 'lift']
rules_clean = rules[important_cols]
rules_clean = rules_clean.sort_values('lift', ascending=False)

rules_clean.head(10)
```
